```python
import numpy as np
import pandas as pd
```

```python
loan_data = pd.read_csv('loan.csv',low_memory=False)
loan_data
```

## Replacing all 0,0.0,null values with nan

```python
loan_data = loan_data.replace([0,0.0,'NULL','n'],np.nan)
loan_data
```

## Dropping coulmns with more than 50% null values

```python
null_percentage = (loan_data.isnull().sum() / len(loan_data)) * 100
columns_to_drop = null_percentage[null_percentage > 50].index
loan_data.drop(columns=columns_to_drop,inplace=True)
loan_data
```

```python
loan_data.dtypes
```

## Dropping some columns which are not useful in analysis

```python
loan_data.drop(['url','desc','initial_list_status','policy_code','emp_title','inq_last_6mths'],axis=1,inplace=True)
loan_data.drop(['total_acc','total_pymnt','total_pymnt_inv','total_rec_prncp','total_rec_int'],axis=1,inplace=True)
loan_data.drop('last_credit_pull_d',axis=1,inplace=True)
loan_data.drop(['title','funded_amnt_inv','dti'],axis=1,inplace=True)
loan_data
```

## Renaming some of column name for better understanding

### Rename based on description given in data dictionary

```python
loan_data.columns = loan_data.columns.str.replace('int_rate', 'interest_rate')
loan_data.columns = loan_data.columns.str.replace('annual_inc', 'annual_income')
loan_data.columns = loan_data.columns.str.replace('emp_length', 'employment_time')
loan_data.columns = loan_data.columns.str.replace('emp_length', 'employment_period')
loan_data.columns = loan_data.columns.str.replace('issue_d', 'loan_issue_month')
loan_data.columns = loan_data.columns.str.replace('addr_state', 'state')
loan_data.columns = loan_data.columns.str.replace('open_acc', 'borrower_open_credit_lines')
loan_data.columns = loan_data.columns.str.replace('revol_util', 'revolving_utilization_rate')
loan_data.columns = loan_data.columns.str.replace('revol_bal', 'revolving_balance')
```

## New Column created based on loan_status name customer_status

* Fully Paid and Current - Non-default
* Charged Off - Default 

```python
loan_data['customer_status'] = loan_data['loan_status'].apply(lambda x : "Default" if x == 'Charged Off' else "Non-default")
```

### Adding the columns "interest_rate_float" and "Salary range" to the loan data filters the data according to the range of interest rates in which each loan's interest rate is contained, same for the salary.

```python
loan_data['interest_rate_float'] = round(loan_data["interest_rate"].str.rstrip("%").astype("float"))
loan_data['interest_rate_float'].head(50)
def range_Distribution(interest):
    if interest < 5:
        return 'less than 5'
    elif 5 <= interest < 10:
        return '5-10'
    elif 10 <= interest < 15:
        return '10-15'
    elif 15 <= interest < 20:
        return '15-20'
    else: 
        return '20-25'
    
loan_data['interest_rate_range'] = loan_data['interest_rate_float'].apply(range_Distribution)
```

```python
loan_data.head(50)
```

```python
Decisions_group=loan_data.groupby('customer_status').get_group('Default')
Interest_graph=Decisions_group.interest_rate_range.value_counts().values
Interest_Range = loan_data['interest_rate_range'].unique()
plt.bar(Interest_Range,Interest_graph)
plt.show()
```

* A bar chart study reveals that customers who are in interest rates of 10-15% are more likely to default on their loan obligations. It results in a 20–25 percent interest rate range with a lower client default rate.

```python
def range_Distribution(Income):
    if Income < 100000:
        return 'less than 1 lac'
    elif 100000 <= Income < 200000:
        return '1-2 lacs'
    elif 200000 <= Income < 300000:
        return '2-3 lacs'
    elif 300000 <= Income < 400000:
        return '3-4 lacs'
    else: 
        return '4-5 lacs'
    
loan_data['Salary_range'] = loan_data['annual_income'].apply(range_Distribution)
```

```python
loan_data.head(50)
```

```python
Decisions_group=loan_data.groupby('customer_status').get_group('Default')
Salary_graph=Decisions_group.Salary_range.value_counts().values
Salary_Range = loan_data['Salary_range'].unique()
plt.bar(Salary_Range,Salary_graph)
plt.show()
```

* Customers who make less than $1 lac in annual income are more likely to default.
* There have been no defaults in the 3–4 lac range.

## Diagram showing the relationship between "annual income" and "interest_rate_float" in a haphazard manner.
* Interest rate range on y-axis
* annual_income of cumtomer in x-axis

```python
scatter_diagram=Decisions_group.plot.scatter(x='annual_income',y='interest_rate_float')
```


* The scatter plot reveals that the highest default customers are those with salaries between 0 and 1 lac, with the maximum default customers being those with salaries under 20,000 that too in interest rate 10-17.5%.

### Ananlysis from above charts - 

* The grade, anuual income, and term of the customer have a significant impact on whether they can repay the entire loan amount.
* Consumers with fewer loan payments (term)are more likely to default than those with more payments(term) (perhaps because they make smaller monthly payments over a longer term).
* The annual income and interest rate have a significant impact on a customer's ability to make loan payments since customers with lower salaries but higher interest rates are more likely to default than those with higher salaries.
